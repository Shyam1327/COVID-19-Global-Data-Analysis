```python
import pandas as pd
confirmed_cases=pd.read_csv('covid_19_confirmed.csv')
confirmed_cases
```

```python
confirmed_cases.shape
```

```python
deaths_cases=pd.read_csv('covid_19_deaths.csv')
deaths_cases
```

```python
recovered_cases=pd.read_csv('covid_19_recovered.csv')
recovered_cases
```

```python
#promoting headers
deaths = pd.read_csv('covid_19_deaths.csv', header=1)
deaths
```

```python
#promoting headers for recovered cases in another way rather than that of deaths promoting headers
recovered_cases.columns
```

```python
recovered_cases.iloc[0]
```

```python
# replacing columns 1 .... column 498 with 1st row which of headers and then dropping that 1st row(other way of promoting headers)
recovered_cases.columns = recovered_cases.iloc[0]
recovered_cases = recovered_cases.drop(0).reset_index(drop =True)
recovered_cases
```

```python
#unpivoting data from wider columns to make it rows (melt)
#pandas melt
#pandas.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None, ignore_index=True)
all_cols=confirmed_cases.columns
all_cols
```

```python
fixed_cols=all_cols[:4]
fixed_cols
```

```python
date_cols=[a for a in all_cols if a not in fixed_cols] # one way of getting date columns
date_cols
```

```python
#other way of getting date cols
date_cols=all_cols[4:]
date_cols
```

```python
#now converting date cols to row(unpivoting )
#pandas.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None, ignore_index=True)
confirmed_long = confirmed_cases.melt(
    id_vars=fixed_cols,  
    value_vars=date_cols,
    var_name="Date",
    value_name='Confirmed'
)
confirmed_long
```

```python
deaths_long = deaths.melt(
    id_vars=fixed_cols,  
    value_vars=date_cols,
    var_name="Date",
    value_name='Deaths'
)
deaths_long
```

```python
recovered_long = recovered_cases.melt(
    id_vars=fixed_cols,  
    value_vars=date_cols,
    var_name="Date",
    value_name='Recovered'
)
recovered_long
```

```python
#structure insigths for each Data Frame
confirmed_long.info()
```

```python
#identifying missing values and cleaning
confirmed_long['Province/State'].isna().sum()
```

```python
#checking values including Nan Values default dropna will be true which will not include Nan values
confirmed_long['Province/State'].value_counts(dropna=False)
```

```python
#Handling Nan values
confirmed_long['Province/State'] = confirmed_long['Province/State'].fillna('All Provinces')
confirmed_long
```

```python
confirmed_long.info()
```

```python
#now handling remaining cols
#since lat and long are based country location we cannot use ffill or bfill because they can be different so replacing with 0
confirmed_long['Lat'] =pd.to_numeric(confirmed_long['Lat'].fillna(0))
confirmed_long['Long'] =pd.to_numeric(confirmed_long['Long'].fillna(0))
confirmed_long
```

```python
confirmed_long.info()
```

```python
#change date data type to date format
confirmed_long['Date'] = pd.to_datetime(confirmed_long['Date'], format = "%m/%d/%y" , errors='coerce')
confirmed_long.info()
```

```python
confirmed_long
```

```python
#Handling and cleaning remaining df
deaths_long.info()
```

```python
# Deaths_long cleaning and handing
deaths_long['Province/State'] = deaths_long['Province/State'].fillna('All Provinces')
deaths_long['Lat'] = pd.to_numeric(deaths_long['Lat'].fillna(0))
deaths_long['Long'] = pd.to_numeric(deaths_long['Long'].fillna(0))
deaths_long['Date'] = pd.to_datetime(deaths_long['Date'], format ="%m/%d/%y", errors='coerce')
deaths_long['Deaths'] = pd.to_numeric(deaths_long['Deaths'].fillna(0).astype('int64'))
deaths_long.info()
```

```python
deaths_long
```

```python
recovered_long.info()
```

```python
# Recovered_long cleaning and handing
recovered_long['Province/State'] = recovered_long['Province/State'].fillna('All Provinces')
recovered_long['Lat'] = pd.to_numeric(recovered_long['Lat'].fillna(0))
recovered_long['Long'] = pd.to_numeric(recovered_long['Long'].fillna(0))
recovered_long['Date'] = pd.to_datetime(recovered_long['Date'], format ="%m/%d/%y", errors='coerce')
recovered_long['Recovered'] = pd.to_numeric(recovered_long['Recovered'].fillna(0).astype('int64'))
recovered_long.info()
```

```python
# 5.1 Analyze the peak number of daily new cases in Germany, France, and Italy. 
# Which country experienced the highest single-day surge, and when did it occur?

target_countries = ['Germany' , 'France' , 'Italy']
country_date_total = (
    merged.loc[merged['Country/Region'].isin(target_countries)]
    .groupby(['Country/Region','Date'])['Confirmed'].sum().reset_index()
    .sort_values(['Country/Region','Date'])
)
country_date_total
```

```python
# Cumulative Confirmed Cases Convert -> Daily New Confirmed Cases [.diff]
country_date_total['DailyNew'] = (
    country_date_total.groupby('Country/Region')['Confirmed'].diff()
)
country_date_total
```

```python
# Identifying the missing Values
country_date_total[country_date_total['DailyNew'].isna()]
```

```python
country_date_total['DailyNew'] = (
    country_date_total.groupby('Country/Region')['Confirmed'].diff()
    .fillna(country_date_total['Confirmed'])
)
country_date_total
```

```python
# Pivot Table
daily_new_matrix = (
    country_date_total.pivot(
        index = 'Date',
        columns = 'Country/Region',
        values = 'DailyNew'
    ).sort_index()
)
daily_new_matrix
```

```python
daily_new_matrix['France'].nlargest(5)
```

```python
daily_new_matrix['Germany'].nlargest(5)
```

```python
daily_new_matrix['Italy'].nlargest(5)
```

```python
daily_new_by_country = {}

# target_countries = ['Germany' , 'France' , 'Italy']

for country in target_countries:
    if country in daily_new_matrix.columns:
        s = daily_new_matrix[country]
        daily_new_by_country[country] = s

daily_new_by_country
```

```python
daily_new_matrix['Germany']
```

```python
daily_new_matrix['France']
```

```python
daily_new_matrix['Italy']
```

```python
# Peak {Country : (max_value , date} per date
peaks = {}
for country , series in daily_new_by_country.items():
    max_value = int(series.max())
    max_index = series.idxmax()
    peaks[country] = (max_value , max_index)

peaks
```

```python
# Visualizations
plt.figure(figsize = (14,8) , facecolor = '#E8E8E8')
for country , series in daily_new_by_country.items():
    plt.plot(series.index , series.values , label = country)

plt.title('Daily New Confirmed Cases For Germany, France , Italy.' , color = 'red')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.ylim(-200000 , 200000)
plt.legend()
plt.show()
```

```python
#7.2 Analyze the monthly sum of confirmed cases, deaths, and recoveries for countries 
# to understand the progression of the pandemic. [From the merged dataset] we make dt.to_perod for monthly values(check datetime documentary)
monthly_data=merged.copy()
monthly_data=monthly_data.sort_values(['Country/Region','Date'])
monthly_data['YearMonth']=monthly_data['Date'].dt.to_period('M')
monthly_data
```

```python
#daily calcaulations of the values
#making use of clip for making limits of the values lower to zero is minimum 0 so that it cannot show negative values
#daily confirmed
dailyconfirm=monthly_data.groupby('Country/Region')['Confirmed'].diff().clip(lower=0)
monthly_data['DailyConfirmed']=dailyconfirm.fillna(0)

#daily Deaths
dailydeath=monthly_data.groupby('Country/Region')['Deaths'].diff().clip(lower=0)
monthly_data['DailyDeaths']=dailydeath.fillna(0)

#daily recovered
dailyrecovered=monthly_data.groupby('Country/Region')['Recovered'].diff().clip(lower=0)
monthly_data['DailyRecovered']=dailyrecovered.fillna(0)

monthly_data
```

```python
monthly_data.info()
```

```python
monthly_trend=(
    monthly_data.groupby(['Country/Region','YearMonth'])[['DailyConfirmed','DailyDeaths','DailyRecovered']]
    .sum().rename(
        columns={
            'DailyConfirmed':'Confirmed',
            'DailyDeaths':'Deaths',
            'DailyRecovered':'Recovered'
        }).astype('int64')
    .reset_index().sort_values(['Country/Region','YearMonth'])
)
monthly_trend
```
